```python
import os


output_base = r'D:\Projects\unified_multimodal_embeddings\audios'


for folder in os.listdir(output_base):
    for file in os.listdir(os.path.join(output_base, folder)):
        print(file)
    break
```

```python
from collections import defaultdict

class_files = defaultdict(list)
```

```python
from pydub import AudioSegment
# NOTE: each class has 40 audio files (somehow those 40 audio files are categorized into 5 types (they start with 1,2,3,4 & 5) )
# step1: merge them
# so by the end.. in each class there will be only 5 audios
# create a dictonary of class id and the 8 audio files which belong to that id (full path).. do that for all 50 classes

for i, class_folder in enumerate(os.listdir(output_base)):
    class_files = defaultdict(list)
    for file in os.listdir(os.path.join(output_base, class_folder)):
        if file.endswith('.wav'):
            class_id = file.split('-')[0]
            class_files[class_id].append(os.path.join(output_base, class_folder, file))

    for class_id, file_list in class_files.items():
        merged = AudioSegment.empty()
        for file_path in sorted(file_list):
            audio = AudioSegment.from_wav(file_path)
            merged += audio
        output_path = os.path.join(output_base, class_folder, f'{class_folder}_{class_id}.wav')
        merged.export(output_path, format='wav')

    print(f"{i+1}: Successfully merged audios for the class {class_folder}")
```

```python
for class_folder in os.listdir(output_base):
    for file in os.listdir(os.path.join(output_base, class_folder)):
        if file[0] in ["1", "2", "3", "4", "5"]:
            os.remove(os.path.join(output_base, class_folder, file))
```
