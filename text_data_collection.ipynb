```python
import torch
from PIL import Image
from transformers import AutoModel, AutoTokenizer

model = AutoModel.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5-int4', trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5-int4', trust_remote_code=True)
```

```python
descriptions = {}
```

```python
!unzip /content/drive/MyDrive/multimodaldata/images.zip -d /content/drive/MyDrive/multimodaldata/images
```

```python
image_base_dir = "/content/drive/MyDrive/multimodaldata/images/images"
```

```python
def get_desc(path):
    model.eval()
    image = Image.open(path).convert('RGB')
    question = 'Give me a comprehensive description of this image'
    msgs = [{'role': 'user', 'content': question}]

    result = model.chat(
        image=image,
        msgs=msgs,
        tokenizer=tokenizer,
        sampling=True,
        temperature=0.7,
    )
    return result
```

```python
%%time
import os
for img_class_folder in os.listdir(image_base_dir):
    full_path = os.path.join(image_base_dir, img_class_folder)
    descriptions[img_class_folder] = {}
    for image in os.listdir(full_path):
        img_desc = get_desc(os.path.join(full_path, image))
        descriptions[img_class_folder][image] = img_desc
        print(f"Description saved for image: {image}")
```

```python
descriptions.keys()
```

```python
new_dir = "/content/drive/MyDrive/multimodaldata/text"
for folder in descriptions:
    class_folder_path = os.path.join(new_dir, folder)
    os.mkdir(class_folder_path)
    for image, text in descriptions[folder].items():
        with open(os.path.join(class_folder_path, f"{image}.txt"), "w") as fout:
            text = text.replace("This image", "This")
            text = text.replace("The image", "This")
            fout.write(text)
```
